# Bar plot


**  A Bar Plot is used to compare different categories.

Think of each bar as representing one category, and the height of the bar represents its value.

Example

Suppose five students scored these marks:

Student	Marks
A	80
B	90
C	75
D	95
E	85

A bar plot makes it easy to see who scored the highest and lowest.

Visual Representation
Marks

100 |
 95 |          █
 90 |    █     █
 85 |    █     █     █
 80 | █  █     █     █
 75 | █  █  █  █     █
     ----------------------
       A  B  C  D  E
**

## Remember


```python
Understanding bar() vs barh()

Vertical Bar Plot
plt.bar(x, height)
Height
 ^
 |
 |      █
 |  █   █
 |  █   █
 +--------------
    A   B
x → Categories
height → Height of bars
width → Thickness of bars


Horizontal Bar Plot
plt.barh(y, width)
A ██████████
B ███████████████
C ███████
--------------------> Width
y → Categories
width → Length of bars (your marks)
height → Thickness of bars

Notice the meaning changes:

In bar(), width means bar thickness.
In barh(), the second positional argument (marks) is the width (length) of each horizontal bar.
```

# structure

```python
plt.bar(x,y)  # remember one thing , the bar plot is used for categorical data

x -> categories
y -> values
```

## Bar plot implementation


```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))

subject = ['maths','english','science','social']
marks = [60,74,85,65]

plt.bar(subject,marks)

plt.title("BAR PLOT")

plt.xlabel(" SUBJECTS ")

plt.ylabel(" MARKS ")

plt.show()
```

## custimization

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))

subject = ['maths','english','science','social']
marks = [60,74,85,65]

plt.bar(subject,
        marks,
        width =0.5,
        color = 'red',
        edgecolor = 'black',
        label = "Bar Plot",
        linewidth =2
       )
font = { 'color' : 'black',
         'fontsize' : 30
       }     

plt.title("BAR PLOT" , fontdict = font)

plt.xlabel(" SUBJECTS ",fontdict = font)

plt.ylabel(" MARKS ",fontdict =font)

plt.grid(axis ='y')

plt.legend(loc = "upper left")

plt.show()
```

# Horizontal Graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))

subject = ['maths','english','science','social']
marks = [60,74,85,65]

plt.barh(
         subject,
         marks,
        height =0.5,
        color = 'orange',
        edgecolor = 'black',
        label = "Bar Plot",
        linewidth =2
       )
font = { 'color' : 'black',
         'fontsize' : 30
       }     

plt.title("BAR PLOT" , fontdict = font)

plt.xlabel(" SUBJECTS ",fontdict = font)

plt.ylabel(" MARKS ",fontdict =font)

plt.grid(axis ='x')

plt.legend(loc = "upper right")

plt.show()
```
