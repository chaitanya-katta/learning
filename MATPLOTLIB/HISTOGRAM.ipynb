# HISTOGRAM

**
A histogram is a graph used to show the distribution of numerical (continuous) data. It groups data into intervals (called bins) and displays the frequency (count) of values in each interval.

Example

Suppose you have the following test scores:

45, 50, 52, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90

A histogram might look like this:

40–49 | *
50–59 | ***
60–69 | ****
70–79 | ****
80–89 | **
90–99 | *

Here:

The x-axis represents score intervals (bins).
The y-axis represents the frequency (number of observations).
Characteristics of a Histogram
Bars touch each other because the data are continuous.
It helps identify:
The shape of the distribution (normal, skewed, uniform, etc.)
The center of the data
The spread (variability)
Outliers or unusual patterns
Histogram vs. Bar Chart
Histogram	Bar Chart
Used for continuous numerical data	Used for categorical data
Bars touch each other	Bars are separated by spaces
Shows frequency distribution	Compares categories
**

** Parameters You Should Know for Interviews & Practice
Parameter	Purpose
x	Input numerical data
bins	Number of intervals
range	Restrict data range
density	Show probability density instead of frequency
weights	Assign custom weights
cumulative	Show cumulative frequencies
histtype	Histogram style (bar, step, etc.)
orientation	Vertical or horizontal bars
rwidth	Relative width of bars
color	Fill color
edgecolor	Border color
linewidth	Border thickness
alpha	Transparency
label	Legend label
stacked	Stack multiple histograms
log	Logarithmic y-axis
align	Align bars (left, mid, right)
bottom	Starting position of bars**

## Basic Structure

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8,6))

marks = [45, 50, 52, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90]

plt.hist(marks,
         bins = 6
        )

plt.title("HISTOGRAM")

plt.xlabel("MARKS")

plt.ylabel("FREQUENCY")

plt.show()
```

# CUSTIMIZATION


```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))

marks = [45, 50, 52, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90]

plt.hist(
         marks,
         bins = 6,
         color = "red",
         edgecolor = "black",
         label = "MARKS",
         linewidth = 3,
         alpha = 0.8,
         rwidth = 1,
         density = True
       )

font = { 'fontsize' : 30,
          'color' : 'black',
       }

plt.title(" HISTOGRAM " , fontdict = font)

plt.xlabel( "MARKS ", fontdict = font)

plt.ylabel("FREQUENCY ",fontdict =font)

plt.legend()


plt.show()
         
```

# Horizonatal histogram

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))

marks = [45, 50, 52, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90]

plt.hist(
         marks,
         bins = 6,
         color = "yellow",
         edgecolor = "black",
         label = "MARKS",
         linewidth = 3,
         alpha = 0.8,
         rwidth = 1,
         orientation = 'horizontal'
       )

font = { 'fontsize' : 30,
          'color' : 'black',
       }

plt.title(" HISTOGRAM " , fontdict = font)

plt.xlabel( "MARKS ", fontdict = font)

plt.ylabel("FREQUENCY ",fontdict =font)

plt.legend()


plt.show()
         
```

# Two Histogram (stacked)

```python
import matplotlib.pyplot as plt

classA = [45,50,55,60,65,70,75]
classB = [50,55,60,65,70,75,80]

plt.hist([classA, classB],
         stacked = True,
         bins=5,
         color=['skyblue', 'orange'],
         edgecolor='black',
         label=['Class A', 'Class B'])

plt.legend()
plt.show()
```
