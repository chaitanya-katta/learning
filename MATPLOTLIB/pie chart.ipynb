# PIE CHART

** 
A pie chart shows how each category contributes to the whole.

Imagine you have a pizza 🍕.

The whole pizza represents 100%.

Each slice represents a portion of that 100%.

For example:

Pepperoni = 40%
Cheese = 30%
Veg = 20%
Mushroom = 10%

The slices together make 100%.

When should you use a Pie Chart?

Ask yourself this question:

"Do I want to know how much each category contributes to the total?"

If the answer is Yes, use a pie chart.
**

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6))

labels = ['AIML', 'CSE', 'ECE', 'EEE']

students = [40, 30, 20, 10]

plt.pie(students,labels= labels)

plt.show()
```

# Custimization

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(10,6))

company = ['Apple','Samsung','Xiami','Others']
share = [45,30,15,10]

font = {
           'fontsize':25,
            'color' : 'black',
            'fontweight':'bold'
       }
plt.pie(
        share,
        labels = company,
        colors = ['green','red','yellow','blue'],
        autopct = '%1.1f%%',
        startangle = 90,
        shadow = True,
        explode = [0.1,0.0,0.0,0.0]
       wedgeprops ={
                      'edgecolor' : 'black'
                 }
       
      )

plt.title("PIE CHart ", fontdict = font)

plt.legend(loc = 'upper right')

plt.show()
         
```
