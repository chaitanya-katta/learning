# Matplotlib 


## Basic Structure

```python
#import statement

import matplotlib.pyplot as plt  

# input values 

x = [1,2,3,4,5]
y = [2,4,6,8,10]

# create the plot dynamically
plt.plot(x,y)

# output / display the plot on the screen
plt.show()
```

## Figure size

```python
plt.figure(figsize = (10,10))  # figure size (width, height)
```

# Line Plot

```python
import matplotlib.pyplot as plt


#define the size

plt.figure(figsize =(10,10))

# input values 

x=[1,2,3,4,5]
y=[5,10,15,20,25]


plt.plot(x,y)

# title of the plot
plt.title("LINE PLOT")

# Naming for the x-axis
plt.xlabel("X Axis")

# Naming for the y-axis
plt.ylabel("Y Axis")

#visulize
plt.show()
```

# custimization

```python

import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.figure(figsize =(10,8))

font = {
         'color' : 'black',
         'fontsize' : 30
      }

#Line 1

plt.plot(x,y,
         marker = "*",
         markersize = 10,
         linewidth = 3,
         linestyle = "--",
         color = 'red',
         label = 'Line Plot'

        )

# Line 2 
a = [1,2,3,4,5]
b= [6,12,18,24,30]
plt.plot(a,b,
         marker = 'o',
         markersize =10,
         label = "Line 2",
         linewidth = 3,
         color = 'black',
        )

plt.title("LINE PLOT ", fontdict = font)

plt.xlabel("X LABEL ", fontdict = font)

plt.ylabel("Y LABEL ", fontdict = font)

plt.legend(loc = 'upper left')

plt.grid()
plt.show()
```
