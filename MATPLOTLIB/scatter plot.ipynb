# Scatter plot


** What is a Scatter Plot?

A scatter plot shows the relationship between two numerical variables.

Think of it as answering the question:

"If one value changes, what happens to the other?"

For example:

Hours Studied	Marks
2	45
3	55
4	65
5	75
6	88**

```python
import matplotlib.pyplot as plt


plt.figure(figsize =(10,8))

hours = [2,3,4,5,6]
marks = [45,55,65,75,88]

plt.scatter(marks,hours)

plt.title( "HOURS VS MARKS ")

plt.xlabel("Hours ")

plt.ylabel("Marks")

plt.show()
```

# Customization

```python


import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

age = [20,22,25,28,30]
salary = [25000,28000,35000,42000,50000]


plt.scatter(age,salary,
            color = 'blue',
            marker = 'o',
            s=200,
            label = 'SCATTER PLOT',
            edgecolors = 'black'
           )

font = {
          'fontsize' : 30,
           'color' : 'black',
           'fontweight': 'bold'
        }

plt.title("AGE VS SALARY ", fontdict = font)

plt.xlabel("AGE ",fontdict = font)

plt.ylabel("SALARY ",fontdict = font)

plt.grid()

plt.legend(loc = 'upper left')

plt.show()
```
