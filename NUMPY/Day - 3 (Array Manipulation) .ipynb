# ARRAY MANIPULATION

```python
# SHAPE OF THE ARRAY 

import numpy as np

arr = np.arange(12) # arange starts from 0 and end before

print(arr)
```

```python
# shape of the array 

print(arr.shape)  # 12 rows , 0 cols


```

## reshape()

```python
# machine learning models did not understand this patter( above code)  , 

# so we change the dimesions not the data , 

#reshape function is used to change the dimensions of the data , not the change in acutual data

#syntax arr.reshape(rows,cols)


print(arr.reshape(3,4))

#Note :- this function work only

'''12 elements

3 × 4 = 12 ✅

2 × 6 = 12 ✅

4 × 3 = 12 ✅

5 × 3 = 15 ❌
'''
```

```python
#example for invalid case

shape = arr.reshape(5,3)

print(shape)
```

## Automatic Dimension (-1)

```python
# numpy calculate the one dimension automatically by using 9-1)

import numpy as np

matrix = np.arange(12)

print(matrix)

# reshaping 
print()
print("After reshaping")
shape = matrix.reshape(3,-1)
print()
print(shape)

```

```python
# working is \
# here the matrix contains the 12 elements , 
# we specify the rows (3), and columns (-1)

# the column dimnesion is automatically calculate like (12/3) = 4  (rows = 3 , total elements = 12) 
```

## row dimension calculation

```python
shape = matrix.reshape(-1,2)

print(shape)
```

## flatten() 

```python
# the flatten function is used to convert any dimension array to 1 D array

import numpy as np

arr = np.arange(20)

shape = arr.reshape(5,4)

print(shape)
```

```python
# flatter the array

print(shape.flatten())
```

```python
# Note : - The flatten creates the copy , the original array is unchanged

print(shape)
```

## ravel()

```python
# the ravel is used to convert the multidimensional array to 1d array ,
# the ravel modifies the original array 

shape.shape
```

```python
# The array(shape) contains the 5 rows and 4 columns 

print(shape.ravel())
```

```python
# The ravel function modifies the original array

print(shape)
```
