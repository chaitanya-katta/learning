## 1 . Reshaping

```python
# creating a basic array 

import numpy as np

arr = np.array([1,2,3,4,5,6])

print("shape is : ",arr.shape)
print(arr)

```

```python
# reshaping ( changing the dimensions not the data)

re = arr.reshape(2,3)

print( "after reshape  :") 
print(re)


#remember the rows x cols should equal to the total number of elements 

# possible cases :
#(1,6)
#(6,1)
#(2,3)
#(3,2)

#invalid cases

#(2,8) required 8 elements (not possible according our array example(arr))
```

## 2 . Reshaping with -1 

```python
# -1 is automatically calculated the row or cols

#example

import numpy as np

matrix = np.arange(6)

matrix = matrix.reshape(2,-1)  # rows = 2 , cols = 6/2 = 3
 
print(matrix)
```

```python
# Example for automatically creating the rows dimension 

mat = matrix.flatten()

mat = mat.reshape(-1,2)  #rows = 6/2 = 3, cols = 2

print(mat)
```

# 3. Row vector vs Column vector

## Row vector

```python
# create a practice array 

import numpy as np

arr = np.arange(1,4)

print(arr)  # it is a 1*3 vector but is shape shows as (3,)


print(arr.shape)  # it creates (3,) this is not same as 1*3 

```

# 4. flatten() vs ravel()

## flatten()

** flatten()** converts the n dimension array into the 1 dimension array

```python
# practice array 

matrix = np.arange(1,13)

matrix = matrix.reshape(4,3)

print(matrix)
```

```python
# let convert the matrix in to 1 d array 

# remember , the flatten creates the copy , it does not effect the original array 

f = matrix.flatten()

print(f)

print()

print("original matrix : ")
print()
print(matrix)
```

```python
# modifying the flatten array f , does not effect the original array 

f[2] = 99 
print(f)

print()

print("original array ")
print()
print(matrix) # original array remains same
```

## ravel()

```python
# ravel() converts an n dimension array to 1 dimension 
# ravel() may returns a view if the array in continuous 

# ravel() may return a copy if the array in non continuous

# changing the ravel() array can effect the original array ( same memory usage)
```

```python
print(matrix)

```

```python
# r is the ravel array

r = matrix.ravel()  

print(r)

print("original array : ")
print(matrix)
```

```python
# changing the ravel array(r) effect the original array

r[2] = 100

print(r)

print("original array : ")

print()
print(matrix)

# see the difference in the output
```

# 5. transpose — turning rows into columns

```python
# Transpose means converting the rows to column && columns to rows

#practice array 

import numpy as np

arr = np.arange(1,7)

arr = arr.reshape(2,3)

print(arr)

print("shape of the array : ")

print(arr.shape)
```

```python
# Apply the transpose 

trans = arr.T

print(trans)

print("shape of the arr after the transpose ", trans.shape)
```

# 7. concatenate() — join arrays

```python
# concatenate()  means merging / joining arrays 

#Rule

#axis=0 → add more rows

#axis=1 → add more columns

# creating two practice arrays 

A = np.arange(1,5).reshape(2,2)

print(A)
print()
B = np.arange(5,9).reshape(2,2)
print(B)
```

```python
# concating axis = 0  (row wise) / vertically
C = np.concatenate((A,B),axis = 0)

C
```

```python
# concaneting axis =1 (column wise) / horizontally

D = np.concatenate((A,B),axis =1)

D
```

# 8. hstack and vstack (easier syntax)

```python
# horizontally stack (column wise)

np.hstack((A,B))
```

```python
# vertically stack ( row wise)

np.vstack((A,B))
```
