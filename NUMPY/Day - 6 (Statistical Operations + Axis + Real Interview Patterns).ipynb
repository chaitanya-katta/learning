# Statistical Operations + Axis + Real Interview Patterns


## 1. Create a sample dataset


```python
# creating the practice array

import numpy as np

arr = np.array([
                 [85,92,105],
                 [88,95,102],
                 [83,98,107]
               ])

print(arr)
```

## 2. Mean()

```python
# mean is the average of the all values in the arr

print(arr.mean())

# mean = sum of all / no.of all
```

## 3.Axis 

```python
# axis = 0 (column wise)

#finding the mean column wise

col = np.mean(arr,axis=0)

# mean for entire each column
print(col)
```

```python
# axis = 1 (row wise)

# finding the mean row wise (1st row, 2 nd row, .....)

row = np.mean(arr,axis=1)

print(row)
```

## 4.sum()

```python
# sum of the entire array 

sum = np.sum(arr)

# all elements 
print(sum)
```

```python
# sum of each column in the array

sum = np.sum(arr,axis = 0)

print(sum)
```

```python
# sum of each row in the array

sum = np.sum(arr,axis =1)

print(sum)
```

## 5. finding the minimum and maximum

```python
# finding the maximum element in the array

min = np.min(arr)

print(min)
```

```python
# finding the maximum element in the array 

max = np.max(arr)

print(max)
```

```python
# finding the max element for each row

max = np.max(arr,axis = 1)

print(max)
```

```python
# finding the min element for each row

min = np.min(arr, axis = 1)

print(min)
```

```python
# finding the min element for each column

min = np.min(arr,axis=0)

print(min)
```

```python
# finding the max element for each column

max = np.max(arr,axis=0)

print(max)
```

## 6.Argmax and Argmin

```python
# this two returns the index not the value

min_index = np.argmin(arr)

print(min_index)
```

```python
# max value index by using argmax

max_index = np.argmax(arr)

print(max_index)
```

## 7. Standard Deviation

```python
# standard deviation means , how the values far from the mean

# low sd means values are close to the mean

# high sd maens values are far from the mean

sd = np.std(arr)

print(sd)
```

## 8. variance

```python
# variance means , how the values are spread out

var = np.var(arr)

print(var)
```

## 9. Median


```python
# median is robust to outliers

median = np.median(arr)

print(median)
```

## 10. Percentiles


```python
# A percentile tells us the position of a value relative to the entire dataset.

p25 = np.percentile(arr,25)

print(p25)
```

```python
# p75

p75 = np.percentile(arr,75)

print(p75)
```

## keepdims()

```python
# keepdims are important in the broadcasting

mean = np.mean(arr,axis=0,keepdims=True)

print(mean)
```

```python
marks = np.array([80, 90, 100])
weights = np.array([1, 2, 3])

np.average(marks, weights=weights)
```
